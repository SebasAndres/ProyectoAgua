```python
## Consumo por registro
import numpy as np

def C(reg):
    
    def tipo_ducha (valor):
        if valor == 1:
            return 250 / 14
        else:
            return 250 / 4 
    
    def tipo_lv (valor):
        if valor == 1:
            return 20
        elif valor == 2:
            return 10
        else:
            return 5
    
    def avoid_nan(x):
        if math.isnan(x):
            return 0
        else: 
            return x
    
    def reciclaje(valor):
        if valor == 1:
            return -20
        elif valor == 2:
            return 0
        else:
            return -50   
    
    AGUA_DUCHA = dc(8)[reg] * tipo_ducha (dc(9)[reg])
    AGUA_LAVABO = dc(10)[reg] * tipo_lv(dc(11)[reg])
    desc_AGUAS_GRISES = (dc(12)[reg]-1) * (-60)
    desc_BARRIL_PLUVIAL = dc(13)[reg] * (-111)
    AGUA_AUTO = dc(17)[reg] * avoid_nan (dc(18)[reg]) * 0.5
    ep = lambda x: (abs(x) + x) / (2 * x + 0.01)
    AGUA_PILETA = ep(dc(19)[reg])*(100 / (dc(19)[reg]+0.00000001)) 
    AGUA_RECICLAJE = reciclaje(dc(20)[reg]) 
    AGUA_CARNE = dc(23)[reg] * 60
    
    return avoid_nan(AGUA_DUCHA) + avoid_nan(AGUA_LAVABO) + avoid_nan(desc_AGUAS_GRISES) + avoid_nan(desc_BARRIL_PLUVIAL) + avoid_nan(AGUA_AUTO) + avoid_nan(AGUA_PILETA) + avoid_nan(AGUA_RECICLAJE) + avoid_nan(AGUA_CARNE)
```

```python
import math

Î¼ = lambda lista: sum(lista)/len(lista)
def Î´(lista, Î¼_):
    f = lambda x: abs(Î¼_ - x)
    return math.sqrt(S(f, len(lista), i=0, datos = lista) / len(lista))

def S(f, k, i=0, datos=[]):
    Accum = 0
    if len(datos) == 0:   
        while i <= k:
            Accum += f(i)
            i+=1
    else:
        for xi in datos:
            Accum += f(xi)
    return Accum

Cr = [C(r) for r in range(len(datos))]
clean_nan = [x for x in Cr if (math.isnan(x) == False)]

print ("> Î¼(Cr) = ", Î¼(clean_nan))
```

```python
# Pertenece
BAJA = [i for i, d in enumerate(datos_full[26]) if d == 2]
MEDIA = [i for i, d in enumerate(datos_full[26]) if d == 1]
ALTA = [i for i, d in enumerate(datos_full[26]) if d == 3]

def modf(lista):
    nl = list()
    for n in lista:
        if n == 2:
            nl.append(1)
        elif n == 1:
            nl.append(2)
        else:
            nl.append(3)
    return nl

# Creen que contamina mÃ¡s
c_BAJA = modf([datos_full[21][i] for i in BAJA])
c_MEDIA = modf([datos_full[21][i] for i in MEDIA])
c_ALTA = modf([datos_full[21][i] for i in ALTA])

def set_colors (X):
    colores = list()
    for x in X:
        if x in BAJA:
            colores.append("blue")
        elif x in MEDIA:
            colores.append("orange")
        else:
            colores.append("red")
    return colores

X = BAJA + MEDIA + ALTA
colores = set_colors(X)

Y = c_BAJA + c_MEDIA + c_ALTA

plt.scatter(X, Y, c=colores)
plt.xlabel("Registros")
plt.ylabel("Â¿QuiÃ©n creen que contamina +? (Baja, Media, Alta)")
plt.savefig("plot.jpg")

## QuiÃ©n cree que es el que mÃ¡s contamina por clase:
print ("< < QuiÃ©n cree que es el que mÃ¡s contamina por clase > >\n")
print ("> Clase baja (1): ", Î¼(c_BAJA))
print ("> Clase media (2): ", Î¼(c_MEDIA))
try:
    print ("> Clase alta (3): ", Î¼(c_ALTA))
except Exception:
    print ("> No hay registros de la 'clase alta'")
print("\n Media General: ", Î¼(Y))
    
## ConclusiÃ³n, toda la sociedad acuerda en que el segmento mÃ¡s contaminante es el de la clase baja
```
